# Modelos Preentrenados

## Xception

```python
from keras.applications import Xception

Xcep = Xception(weights='imagenet',  #  pesos de inicio del modelo     
                 include_top=True,   #  # incluye el clasificador
                 input_shape=(299, 299, 3))  # dimension de entrada (opcional)
Xcep.summary()
```

## Inception V3

```python
from keras.applications import InceptionV3

Incep = InceptionV3(weights='imagenet',
                 include_top=True,
                 input_shape=(299, 299, 3))
Incep.summary()
```

## ResNet50

```python
from keras.applications import ResNet50

RN = ResNet50(weights='imagenet',
                 include_top=True,
                 input_shape=(224, 224, 3))
RN.summary()
```

## MobileNet

```python
from keras.applications import MobileNet

MN = MobileNet(weights='imagenet',
                 include_top=True,
                 input_shape=(224, 224, 3))
MN.summary()
```

## VGG19

```python
from keras.applications import VGG19

vgg19 = VGG19(weights='imagenet',
              include_top=True,
              input_shape=(224, 224, 3))
vgg19.summary()
```

## VGG16

```python
from keras.applications import VGG16

# red completa
vgg16 = VGG16(weights='imagenet',
              include_top=True,
              input_shape=(224, 224, 3))

vgg16.summary()
```

```python
# red sin el clasificador (flatten-fc1-fc2-prediction)
conv_base = VGG16(weights='imagenet',
                  include_top=False,
                  input_shape=(150, 150, 3))

conv_base.summary()
```
